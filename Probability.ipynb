```python
##Probability Examples for CHM 372
##Goal to cover classic probability problems for the students to play with and learn
###Made on April, 1st 2022 by Jacob Wat
###Last Edited on April 22,2022 by Jacob Wat

#####Examples to include
##### 1. Dice roll unweighted
##### 2. Dice roll weighted
##### 3. Coin flips 
##### 4. Probability distributions
##### 5. Random Walk 1D and 2D walk
##### 6. Weighted Random walks for 1 and 2 D
##### 7. Add program for energy distribution using boltzman



import math
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as random
import re
import time
import ipywidgets as widgets
import matplotlib
from IPython.display import display, clear_output
import tkinter as tk
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,AutoMinorLocator)
import decimal 
from ipywidgets import Box, Layout
from matplotlib.gridspec import GridSpec

def VBox(*pargs, **kwargs):
    """Displays multiple widgets vertically using the flexible box model."""
    box = Box(*pargs, **kwargs)
    box.layout.display = 'flex'
    box.layout.flex_flow = 'column'
    box.layout.align_items = 'stretch'
    return box

def HBox(*pargs, **kwargs):
    """Displays multiple widgets horizontally using the flexible box model."""
    box = Box(*pargs, **kwargs)
    box.layout.display = 'flex'
    box.layout.align_items = 'stretch'
    return box


items_layout = Layout( width='auto')     # override the default width of the button to 'auto' to let the button grow

box_layout = Layout(display='flex',
                    flex_flow='column',
                    align_items='stretch',
                    border='solid',
                    width='50%')
```

```python
############################################################3

## First widget is a list of all available functions for user to choose from
## with rows set to 6 as there are 6 options for now

from ipywidgets import interactive
ProbOptions=widgets.Select(
    options=['Die Roll', 'Dice Roll','Coin Flip','Multicoin Flip', '1D Random Walk', '2D Random Walk', "Boltzman Energy?"],
    value='Die Roll',
    rows=7,
    description='Probability:',
    disabled=False
)


###Button is used here to allow for input for the previous widget ot be registered so that the input can be accepted

button = widgets.Button(description="Select!")
output = widgets.Output()
display(ProbOptions,button, output)

##   Widget functions here are selected based on a series of if statements bassed on the string produced by the list selection

def on_button_clicked(b):
    with output:
        clear_output()
        
        #############################################################################################
        #############################################################################################
        #####  Die Roll Simulation
        #####  ---Goal: Students will be able to see how probability works for a simple Die Roll
        
        if ProbOptions.value == 'Die Roll':  
            NumDie=widgets.FloatSlider(
                value=7,
                min=0,
                max=100,
                step=1,
                description='Number of Dice:',
                disabled=False,
                continuous_update=True,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            buttonDieRoll = widgets.Button(description="Click Me!")
            outputDieRoll = widgets.Output()
            display(VBox([HBox([NumDie,buttonDieRoll]), outputDieRoll]))

            def on_button_clicked2(c):
                with outputDieRoll:
                    clear_output()
                    dice_roll=[0,0,0,0,0,0]
                    record=[]
                    for n in range (0,int(NumDie.value)):
                        clear_output(wait=True)
                        fig=plt.figure(figsize=(12,12))
                        plt.axes().set_aspect('auto')
                        number=np.random.randint(0,6)
                        record.append(number)
                        dice_roll[number]+=1
                        plt.bar([1,2,3,4,5,6],dice_roll)                     
            
                        plt.gca().axes.yaxis.set_visible(False) 
                        plt.xticks(fontsize=30)

                        plt.title(str(int(NumDie.value))+" Dice Rolls"+": Die roll "+str(n+1),fontsize=30)
                        plt.xlabel(xlabel=("Die Value"),fontsize=40)
                        plt.show()
                    clear_output(wait=True)
                    fig=plt.figure(figsize=(12,12))
                    plt.bar([1,2,3,4,5,6],dice_roll/np.max(dice_roll))
                    plt.gca().axes.yaxis.set_visible(False) 
                    plt.xticks(fontsize=30)
                    plt.xlabel(xlabel=("Die Value"),fontsize=40)

                    plt.title(str(int(NumDie.value))+" Die Rolls"+": Die roll "+str(n+1),fontsize=30)
                    for n in range (0,6):
                        plt.text(n+0.9,0.25,dice_roll[n],fontsize=30)
                    average=np.average(record)
                    median=np.median(record)
                    
        

                    plt.text(7,0.6,"The Average Value is: "+str(np.round(average,2)),fontsize=20)
                    plt.text(7,0.5,"The Median Value is: "+str(median),fontsize=20)

                    plt.show()
                    #print("Average Value: "+str(np.round(average,2)))
            buttonDieRoll.on_click(on_button_clicked2)
        #############################################################################################
        #############################################################################################
        #####  Dice Roll Simulation
        #####  ---Goal: Students will be able to how Dice rolls will reach a gaussian shape
        
        if ProbOptions.value == 'Dice Roll': #Dice Roll simulator
            RollsDice=widgets.FloatSlider( # Number of rolls/Trials
                value=20,
                min=1,
                max=200,
                step=1,
                description='Rolls:',
                disabled=False,
                continuous_update=True,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            NumDice=widgets.FloatSlider( #Number of Dice per trial
                value=2,
                min=1,
                max=3,
                step=1,
                description='Dice:',
                disabled=False,
                continuous_update=True,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )



            buttonDiceRoll = widgets.Button(description="Click Me!")
            outputDiceRoll = widgets.Output()

            display(VBox([HBox([VBox([NumDice,RollsDice]),buttonDiceRoll]), outputDiceRoll]))

            def on_button_clickedDice(d):
                with outputDiceRoll:
                    clear_output()
                    dice_roll=[]
                    x_axis=[]
                    for n in range (int(NumDice.value),6*(int(NumDice.value))+1):
                        dice_roll.append(0)
                        x_axis.append(n)
                    record=[]
                   
                    for n in range (0,int(RollsDice.value)):
                        clear_output(wait=True)
                        fig=plt.figure(figsize=(12,12))
                        plt.axes().set_aspect('auto')
                        number=0
                       
                        for m in range(0,int(NumDice.value)):
                            number+=np.random.randint(0,6)
                        
                        record.append(number+2)
                        dice_roll[number]+=1
                        plt.bar(x_axis,dice_roll)    
                        plt.gca().axes.yaxis.set_visible(False) 
                        plt.xticks(fontsize=30)
                        plt.xlabel("Number of times Dice Sum was Rolled", fontsize=35)
                        plt.title(str(int(NumDice.value))+" Dice"+": Dice roll "+str(n+1),fontsize=30)
                        plt.show()
                    clear_output(wait=True)
                    fig=plt.figure(figsize=(12,12))
                    plt.bar(x_axis,dice_roll/np.max(dice_roll))    
                    plt.gca().axes.yaxis.set_visible(False) 
                    plt.xticks(fontsize=30)
                    plt.title(str(int(NumDice.value))+" Dice "+": Dice roll "+str(n+1),fontsize=30)
                    for n in range (0,len(dice_roll)):
                        plt.text(int(NumDice.value)+n-1+0.9,0.025,dice_roll[n],fontsize=30)
                    average=np.average(record)
                    median=np.median(record)
                    mode=np.argmax(dice_roll)+int(NumDice.value)
                    plt.xlabel("Sum of Dice Values", fontsize=35)
                    plt.text(6*(int(NumDice.value))/.9,1,"The Average Value is: "+str(np.round(average,2)),fontsize=20)
                    plt.text(6*(int(NumDice.value))/.9,0.9,"The Median Value is: "+str(median),fontsize=20)
                    plt.text(6*(int(NumDice.value))/.9,0.8,"The Mode Value is: "+str(mode),fontsize=20)
                    plt.show()
                    
                    #print(record)     #These lines are kept for reference if we believe there is an error
                    #print(dice_roll)
            buttonDiceRoll.on_click(on_button_clickedDice)
 

        #############################################################################################
        #############################################################################################
        #####  Coin Flip Simulation
        #####  ---Goal: Students will be able to better visualize how a two option probability will look

        if ProbOptions.value == 'Coin Flip':
            CoinFlips=widgets.FloatSlider(
                value=50,
                min=1,
                max=100,
                step=1,
                description='Flips:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )

            buttonCoinFlip = widgets.Button(description="Click Me!")
            outputCoinFlip = widgets.Output()

            display(HBox([VBox([CoinFlips,buttonCoinFlip]), outputCoinFlip]))

            def on_button_clickedCoinFlip(b):
                with outputCoinFlip:
                    clear_output()
                    heads=0
                    tails=0
                    X_axis=["Heads", "Tails"]
                    for n in range (0,int(CoinFlips.value)):
                        clear_output(wait=True)
                        fig=plt.figure(figsize=(10,10))
                        plt.axes().set_aspect('auto')
                        x=np.random.randint(2)
                        if x == 0:
                            heads+=1
                        if x == 1:
                            tails+=1
                        Coin_Matrix=[heads,tails]
                        Coin_Matrix/=np.max(Coin_Matrix)
                        plt.bar(X_axis,Coin_Matrix)
                        plt.gca().axes.yaxis.set_visible(False)
                        plt.xticks(fontsize=30)
                        
                        plt.title("Trial "+str(int(np.round(n))+1)+" out of "+str(int(np.round(int(CoinFlips.value))))+" Trials", fontsize=30)
                        plt.text(0,0.25,str(int(np.round(heads))),fontsize=40)
                        plt.text(1,0.25,str(int(np.round(tails))),fontsize=40)

                        plt.show()
            buttonCoinFlip.on_click(on_button_clickedCoinFlip)
            
            
        #############################################################################################
        #############################################################################################
        #####  Multi Coin Flip Simulation
        #####  ---Goal: Students will be able to better visualize how a series of coin flips will
        ####            a gaussian shape over time
        if ProbOptions.value == 'Multicoin Flip':
            MultiCoinCoins=widgets.FloatSlider(
                value=50,
                min=1,
                max=100,
                step=1,
                description='Coins:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )

            MultiCoinTrials=widgets.FloatSlider(
                value=50,
                min=1,
                max=100,
                step=1,
                description='Trials:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            MultiCoinbutton = widgets.Button(description="Click Me!")
            MultiCoinoutput = widgets.Output()
            display(MultiCoinCoins,MultiCoinTrials, MultiCoinbutton,MultiCoinoutput)
            def MultiCoinon_button_clicked(b):
                with MultiCoinoutput:
                    clear_output()
                    X_axis=[0]
                    for n in range (0,int(MultiCoinCoins.value)):
                        X_axis.append(n+1)
                    Record=np.zeros(len(X_axis))
                    Num_Record=[]
                    for n in range(0,int(MultiCoinTrials.value)):
                        clear_output(wait=True)
                        Heads=0
                        for m in range(0,int(MultiCoinCoins.value)):
                            MultiCoinx=np.random.randint(0,2)
                            if MultiCoinx == 1:
                                Heads+=1
                        Record[Heads]+=1
                        Num_Record.append(Heads)
                        
                        fig=plt.figure(figsize=(16,12))
                        plt.bar(X_axis,Record/np.max(Record))
                        plt.axes().set_aspect('auto')
                        plt.gca().axes.yaxis.set_visible(False) 
                        plt.xticks(fontsize=30)
                        plt.axes().xaxis.set_minor_locator((MultipleLocator(2)))
                        plt.axes().tick_params(which='both', width=5)
                        plt.axes().tick_params(which='major', length=15)
                        plt.axes().tick_params(which='minor', length=10, color='b')
                        plt.xlabel("Total Number of Trials with N Sum of Heads", fontsize=30)
                        if int(MultiCoinTrials.value) == n+1:
                            average=np.round(np.average(Num_Record),4)
                            average_rounded=np.round(average)
                            median=np.median(Num_Record)
                            mode=np.argwhere(Record == np.amax(Record))
                            modelist=""
                            for m in range (0,len(mode)):
                                modelist+=str(int(mode[m]))+"   "
                            plt.text(0,1,"The Mean is "+str(average),fontsize=20)
                            plt.text(0,0.9,"The Rounded Mean is "+str(average_rounded),fontsize=20)
                            plt.text(0,0.8,"The Median is "+str(median),fontsize=20)
                            plt.text(0,0.7,"The Mode is "+modelist,fontsize=20)
                        plt.title(str(n+1)+" Trials Ran out of "+str(int(MultiCoinTrials.value))+" Trials",fontsize=30)
                        
                        plt.show()
                    #print(X_axis)  ###Extra lines serve as a way to test the data if needed 
                    #print(Record)
            MultiCoinbutton.on_click(MultiCoinon_button_clicked)
            
        #############################################################################################
        #############################################################################################
        #####  One Dimensional Random Wal Simulation
        #####  ---Goal: Students will be able to see a random walk and how weight/favoritism shapes the motion
            
        if ProbOptions.value == "1D Random Walk":###1Dimensional Random Walk
            OneDRandomRalkWeight=widgets.FloatSlider(
                value=0,
                min=-100,
                max=100,
                step=1,
                description='Weight:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )



            OneDRandomRalkSteps=widgets.FloatSlider(
                value=50,
                min=1,
                max=500,
                step=1,
                description='Steps:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )

            OneDRandomRalkbutton = widgets.Button(description="Click Me!")
            OneDRandomRalkoutput = widgets.Output()
            
            
            display(VBox([HBox([VBox([OneDRandomRalkWeight,OneDRandomRalkSteps]),OneDRandomRalkbutton]), OneDRandomRalkoutput]))
            def OneDRandomRalkon_button_clicked(b):
                with OneDRandomRalkoutput:
                    clear_output()
                    x=0
                    plt.ion()
                    xcoord=[]
                    ycoord=[0]
                    xcoord.append(x)
                    for n in range (0,int(OneDRandomRalkSteps.value)):
                        fig=plt.figure(figsize=(16,8))
                        plt.axes().set_aspect('auto')
                        favor=int(OneDRandomRalkWeight.value)+100
                        xstep=np.random.randint(0,200)
                        if xstep <= favor:
                            x+=1
                        else:
                            x-=1
                        xcoord.append(x)
                        ycoord.append(0)
                        clear_output(wait=True)
                        
                        plt.scatter(0,0, c="black",s=100)
                        fig=plt.plot(xcoord,ycoord,linestyle="--")
                        
                        plt.scatter(x,0,c="red",s=100)
                        plt.title("Random walk for "+str(int(OneDRandomRalkSteps.value))+" steps"+": step "+str(n+1),fontsize=30)
                        if n+1==int(OneDRandomRalkSteps.value): #final label at the end of the steps
                            distance=np.round(x,3)
                            average_x=np.round(np.average(xcoord),3)
                            plt.text(np.min(xcoord),0.04,"Average position is "+str(average_x)+", "+str(0),fontsize=30)
                            plt.text(np.min(xcoord),0.03,"Distance from origin: "+str(distance),fontsize=30)
                            
                        
                        plt.gca().axes.yaxis.set_visible(False) 
                        plt.gca().axes.xaxis.set_visible(False) 
                        plt.show()

            OneDRandomRalkbutton.on_click(OneDRandomRalkon_button_clicked)
        #############################################################################################
        #############################################################################################
        ##### 2D Random Walk Simulation
        #####  ---Goal: Students will be able to see how a particle moves in 2D space as an example 
        #####           of movement and how a particle should remain near zero
            
        if ProbOptions.value == "2D Random Walk": ## Two dimensional random walk
            TwoDSteps=widgets.FloatSlider(
                value=50,
                min=1,
                max=500,
                step=1,
                description='Steps:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )

            TwoDbutton = widgets.Button(description="Click Me!")
            TwoDoutput = widgets.Output()

            display(TwoDSteps,TwoDbutton, TwoDoutput)


            def TwoDon_button_clicked(b):
                with TwoDoutput:
                    clear_output()
                    x=0
                    y=0
                    plt.ion()
                    xcoord=[]
                    ycoord=[]
                    xcoord.append(x)
                    ycoord.append(y)
                    for n in range (0,int(TwoDSteps.value)):
                        fig=plt.figure(figsize=(10,10))
                        plt.axes().set_aspect('auto')
                        xstep=np.random.randint(0,2)
                        ystep=np.random.randint(0,2)
                        if xstep == 1:
                            x+=1
                        else:
                            x-=1
                        if ystep == 1:
                            y+=1
                        else:
                            y-=1
                        xcoord.append(x)
                        ycoord.append(y)
                        clear_output(wait=True)
                        plt.scatter(0,0, c="black",s=100)
                        fig=plt.plot(xcoord,ycoord,linestyle="--")
                        plt.scatter(x,y,c="red",s=100)
                        plt.title("Random walk for "+str(int(TwoDSteps.value))+" steps"+": step "+str(n+1),fontsize=30)
                        if n+1==int(TwoDSteps.value):
                            distance=np.round(np.sqrt(x**2+y**2),3)
                            average_x=np.round(np.average(xcoord),3)
                            average_y=np.round(np.average(ycoord),3)
                            plt.text(np.max(xcoord)+np.max(xcoord)*0.15,np.min(ycoord)-0.1*np.min(ycoord),"Average position is "+str(average_x)+", "+str(average_y),fontsize=20)
                            plt.text(np.max(xcoord)+np.max(xcoord)*0.15,np.min(ycoord)-0.3*np.min(ycoord),"Distance from origin: "+str(distance),fontsize=20)
                        plt.gca().axes.yaxis.set_visible(False) 
                        plt.gca().axes.xaxis.set_visible(False) 
                        plt.show()

            TwoDbutton.on_click(TwoDon_button_clicked)
        
        
        #############################################################################################
        #############################################################################################
        ##### Boltzman Energy Simulation
        #####  ---Goal: Students will be able to see how Z' and Energy relate and how the 
        #####           contributions to energy change between each state with time
        
        
        if ProbOptions.value == "Boltzman Energy?":
            
            BD_Temperature=widgets.FloatSlider(
                                value=0,
                                min=0,
                                max=500,
                                step=1,
                                description='Temperature:',
                                disabled=False,
                                continuous_update=False,
                                orientation='horizontal',
                                readout=True,
                                readout_format='.1f',
                            )
            BD_States=widgets.FloatSlider(
                value=3,
                min=2,
                max=3,
                step=1,
                description='Excited States:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            BD_Start = widgets.Button(description="Click Me!")
            BD_output = widgets.Output()
            display(BD_Temperature,BD_States, BD_Start,BD_output)
            
            
            #####################################################
            ######## Degeneracy of each state
            #####################################################
            Deg_0=widgets.FloatSlider(
                value=1,
                min=1,
                max=5,
                step=1,
                description='GS Degeneracy:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            Deg_1=widgets.FloatSlider(
                value=1,
                min=1,
                max=5,
                step=1,
                description='Ex1 Degeneracy:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            Deg_2=widgets.FloatSlider(
                value=1,
                min=1,
                max=5,
                step=1,
                description='Ex2 Degeneracy:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            #####################################################
            ######## Energy of each state relative to 0 J
            #####################################################
            Ex_0=widgets.FloatSlider(
                value=0,
                min=0,
                max=99,
                step=1,
                description='GS Energy:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            Ex_1=widgets.FloatSlider(
                value=10,
                min=100,
                max=199,
                step=1,
                description='Ex1 Energy:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            Ex_2=widgets.FloatSlider(
                value=200,
                min=200,
                max=499,
                step=1,
                description='Ex2 Energy:',
                disabled=False,
                continuous_update=False,
                orientation='horizontal',
                readout=True,
                readout_format='.1f',
            )
            Boltzman=1#1.38*10**(-23)
            T_mat=[]
            for n in range (0,5000):
                T_mat.append(n)
            
            def BD_button_clicked(b):
                with BD_output:
                    N_states=int(BD_States.value)
                    clear_output()
                    N_mat=[]
                    print(N_states)
                    for n in range (0,N_states):
                        N_mat.append(n)
                    if N_states == 2:
                        vbox1=VBox([Deg_0,Deg_1])
                        vbox2=VBox([Ex_0,Ex_1])
                    if N_states == 3:
                        vbox1=VBox([Deg_0,Deg_1,Deg_2])
                        vbox2=VBox([Ex_0,Ex_1,Ex_2])
                    
                    BD2_Start = widgets.Button(description="Click Me!")
                    BD2_output = widgets.Output()
                    display(VBox([HBox([vbox1,vbox2]),BD2_Start,BD2_output]))
                    
                    def BD2_button_clicked(b):
                        with BD2_output:
                            clear_output()
                            #fig, axes =plt.subplots(2,2)
                            fig=plt.figure(figsize=(12,14))
                            gs = GridSpec(nrows=3, ncols=2)
                            #print(N_mat)
                            Z_mat=[]
                            Ex0=float(int(Ex_0.value))
                            Ex1=float(int(Ex_1.value))
                            Ex2=float(int(Ex_2.value))
                            Ex2-=Ex0
                            Ex1-=Ex0
                            Ex0-=Ex0
                            Deg0=float(int(Deg_0.value))
                            Deg1=float(int(Deg_1.value))
                            Deg2=float(int(Deg_0.value))
                            Z_mat=[]
                            En_mat=[]
                            En0_mat=[]
                            En1_mat=[]
                            En2_mat=[]
                            Prob_0=[]
                            Prob_1=[]
                            Prob_2=[]
                
                            for m in range (0,len(T_mat)):
                        
                                if m == 0:
                                    Z=Deg0
                                    avEn=Ex0
                                    Prob_0.append(1)
                                    Prob_1.append(0)
                                    Prob_2.append(0)
                                    
                                if m != 0:
                                    beta=T_mat[m]*Boltzman
                                    e=math.e
                                    #print(np.exp(-Ex0/(beta)))
                                    e0=(Deg0*e**(-Ex0/(beta)))
                                    e1=(Deg1*e**(-Ex1/(beta)))
                                    e2=(Deg2*e**(-Ex2/(beta)))
                                    if N_states == 2:
                                        Z=e0+e1
                                        avEn=(Ex0*e0+Ex1*e1)/(e0+e1)
                                        En_0=(Ex0*e0)/(e0+e1)
                                        En_1=(Ex1*e1)/(e0+e1)
                                        Prob_0.append(e0/Z)
                                        Prob_1.append(e1/Z)
                                    
                                    if N_states == 3:
                                        Z=e0+e1+e2
                                        avEn=(Ex0*e0+Ex1*e1+Ex2*e2)/Z
                                        En_0=(Ex0*e0)/(e0+e1+e2)
                                        En_1=(Ex1*e1)/(e0+e1+e2)
                                        En_2=(Ex2*e2)/(e0+e1+e2)
                                        Prob_0.append(e0/Z)
                                        Prob_1.append(e1/Z)
                                        Prob_2.append(e2/Z)
                                En_mat.append(avEn)
                                Z_mat.append(Z)
                                
                                
                            ######################################################################3
                            ################# Code that developes the figures that are show
                            ################# Program is set to make four plots for the students
                            TvsEn = fig.add_subplot(gs[0,0])
                            TvsEn.set_title("Average Energy as a function of Temperature",fontsize=16)
                            TvsEn.set_xlabel("Temperature (K)",fontsize=20)
                            TvsEn.set_ylabel("Energy (J)",fontsize=20)
                            TvsEn.plot(T_mat,En_mat,linewidth=7,color="green")
                            TvsEn.tick_params(axis='both', which='major', labelsize=15)                            
                            
                            TvsZ = fig.add_subplot(gs[0,1])
                            TvsZ.set_title("Z' as a function of Temperature",fontsize=16)
                            TvsZ.set_xlabel("Temperature (K)",fontsize=20)
                            TvsZ.set_ylabel("Z'",fontsize=20)
                            TvsZ.plot(T_mat,Z_mat,linewidth=7,color="orchid")
                            TvsZ.tick_params(axis='both', which='major', labelsize=15)  
                            
                            
                            TvsP = fig.add_subplot(gs[1:,0:])
                            TvsP.set_title("Percent Contribution as a function of Temperature",fontsize=30)
                            TvsP.set_xlabel("Temperature (K)",fontsize=30)
                            TvsP.set_ylabel("Percent Contribution",fontsize=30)
                            if N_states ==2:
                                TvsP.plot(T_mat,Prob_0,linewidth=7,color="darkblue")
                                TvsP.plot(T_mat,Prob_1,linewidth=7,color="crimson")
                                TvsP.text(3000,0.95,"Ground State, Dark Blue",fontsize=15)
                                TvsP.text(3000,0.88,"First Excited State, Crimson",fontsize=15)
                            if N_states ==3:
                                TvsP.plot(T_mat,Prob_0,linewidth=7,color="darkblue")
                                TvsP.plot(T_mat,Prob_1,linewidth=7,color="crimson")
                                TvsP.plot(T_mat,Prob_2,linewidth=7,color="goldenrod")
                                TvsP.text(3000,0.95,"Ground State, Dark Blue",fontsize=15)
                                TvsP.text(3000,0.88,"First Excited State, Crimson",fontsize=15)
                                TvsP.text(3000,0.81,"Second Excited State, Yellow",fontsize=15)
                            TvsP.tick_params(axis='both', which='major', labelsize=20)  
                   
                            fig.tight_layout()
                            plt.show()
                    BD2_Start.on_click(BD2_button_clicked)
            BD_Start.on_click(BD_button_clicked)
        






                        






button.on_click(on_button_clicked) 
```
